```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("Data File/final_model_data.csv")
```

```python
dataframe = dataframe.rename(columns={'mandeath':'target'})
dataframe = dataframe.drop('Unnamed: 0',axis=1)
dataframe = dataframe.drop('index',axis=1)

y = dataframe['target'].to_frame()
X=dataframe.drop('target',axis=1)
```

```python
for feature in X:
    if feature == 'Sex_F':
        break
    
    df_feature = pd.DataFrame(X,columns=[feature])
    df = pd.concat([df_feature,y],axis=1)
    
    sns.heatmap(df.corr(),annot=True)
    plt.show()

    
```
