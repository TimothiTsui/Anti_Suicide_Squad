```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('values_can_be_trained.csv',low_memory=False)

df[['monthdth']]=df[['monthdth']].astype(int)
df[['year']]=df[['year']].astype(int)
monthly_trends = pd.pivot_table(df, index =['monthdth'],columns=['year'],values='mandeath' ,aggfunc = np.sum)


```

```python
fig = plt.figure(figsize=(12.8,9.6))
plt.plot(monthly_trends)
plt.legend(monthly_trends)
```

```python

monthly_trends.to_csv(r'time_series.csv',index=True)
```
